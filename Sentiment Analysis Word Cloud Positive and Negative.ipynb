Sentiment analysis types and approaches
Sentiment analysis tasks can be carried out at different levels of granularity. 
•First is document level. This is when we look at the whole review of a product, for example. 
•Second is the sentence level. This refers to determining whether the opinion expressed in each sentence is positive, negative, or neutral. 
•The last level of granularity is the aspect level. The aspect refers to expressing opinions about different features of a product. Imagine a sentence such as "The camera in this phone is pretty good but the battery life is disappointing." It expresses both positive and negative opinions about a phone and we might want to be able to say which features of the product clients like and which they don't.

Type of sentiment analysis algorithms
The algorithms used for sentiment analysis could be split into 2 main categories. 
•The first is rule or lexicon based. Such methods most commonly have a predefined list of words with a valence score. For example, nice could be +2, good +1, terrible -3, and so on. The algorithm then matches the words from the lexicon to the words in the text and either sums or averages the scores in some way. As an example, let's take the sentence, 'Today was a good day.' Each word gets a score, and to get the total valence we sum the words. In this case, we have a positive sentence. 
•A second category is automated systems, which are based on machine learning. This is going to be our focus in this course. The task is usually modeled as a classification problem where using some historical data with known sentiment, we need to predict the sentiment of a new piece of text.

Detecting the sentiment of Tale of Two Cities
One type of algorithms for detecting the sentiment are based on a lexicon of predefined words and their corresponding polarity score. In the sentence, we will detect the sentiment, including polarity and subjectivity of a given string (texts) using such a rule-based approach and the textblob library in Python.

We will work with the two_cities string. It contains the first sentence of Dickens's A Tale of Two Cities novel.

```python
two_cities = "It was the best of times, it was the worst of times, it was the age of wisdom, it was the age of foolishness, it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness, it was the spring of hope, it was the winter of despair, we had everything before us, we had nothing before us, we were all going direct to Heaven, we were all going direct the other way—in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only."
# Import the required packages
from textblob import TextBlob

# Create a textblob object 
blob_two_cities = TextBlob(two_cities)

# Print out the sentiment   
print(blob_two_cities.sentiment)
```

Polarity
Polarity measures the emotional content of the text, ranging from -1.0 to +1.0. A negative value indicates a negative sentiment, a positive value indicates a positive sentiment, and a value close to 0 indicates a neutral sentiment.
In the case above, the polarity is approximately 0.023, which is very close to 0. This suggests that the overall sentiment of the text is neutral, with a very slight tilt towards positivity.

Subjectivity
Subjectivity quantifies the amount of personal opinion and factual information contained in the text. The range is from 0.0 to 1.0, where 0.0 is very objective and 1.0 is very subjective.

The subjectivity score output is approximately 0.590. This indicates that the text is moderately subjective, suggesting that it contains a significant amount of personal opinion or personal experience rather than purely factual content.

Overall Interpretation
The text which was analyzed with TextBlob likely expresses a mild personal opinion or perspective rather than strong emotions or clear factual content. The neutrality in polarity alongside moderate subjectivity might suggest balanced or carefully phrased language that includes personal viewpoints or assessments.

```python
# Comparing the sentiment of two strings
# We will compare the sentiment of two different strings. A string called annak that contains the first sentence of Anna Karenina. A second string called catcher  it contains the first sentence of The Catcher in the Rye.


annak = "Happy families are all alike; every unhappy family is unhappy in its own way."
catcher = "If you really want to hear about it, the first thing you'll probably want to know is where I was born, an what my lousy childhood was like, and how my parents were occupied and all before they had me, and all that David Copperfield kind of crap, but I don't feel like going into it, if you want to know the truth."

# annak link address
# http://triggs.djvu.org/djvu-editions.com/TOLSTOY/ANNA/Download.pdf
    
# catcher
# https://simontechnology.org/ourpages/auto/2013/1/23/53406450/Catcher%20in%20the%20Rye%20Text.pdf
    
# Create a textblob object 
blob_annak = TextBlob(annak)
blob_catcher = TextBlob(catcher)

# Print out the sentiment   
print('Sentiment of annak: ', blob_annak.sentiment)
print('Sentiment of catcher: ', blob_catcher.sentiment)
```

Another Example. What is the sentiment of a movie review?
Now we will work with the movie reviews dataset. We will find longest and shortest reviews. One of the longest reviews has been imported for you. It is called titanic as it discusses the Titanic movie. 

In addition, we will calculate the polarity and subjectivity of the titanic string. This review is positive (i.e. has a label of 1). Is the polarity score also positive?

```python
# We will fint the longest review in our movie dataset

# Calculate the length of each review
movies['review_length'] = movies['review'].str.len()

# Find the index of the longest review
index_of_longest_review = movies['review_length'].idxmax()

# Display the longest review
longest_review = movies.loc[index_of_longest_review, 'review']
print("The longest review is:")
print(longest_review)
```

```python
# To conduct a sentiment movie review

import pandas as pd
from textblob import TextBlob

# Add a new column for review length
movies['review_length'] = movies['review'].str.len()

# Find the longest review
longest_review = movies.loc[movies['review_length'] == movies['review_length'].max(), 'review'].iloc[0]

# Perform sentiment analysis on the longest review
blob = TextBlob(longest_review)
print("Sentiment of the longest review: ", blob.sentiment)
```

Polarity
Polarity: 0.2025 (approximately)
This score indicates that the sentiment of the review is moderately positive. Polarity scores range from -1 (very negative) to +1 (very positive), with 0 being neutral. A score of around 0.2025 suggests that the review has a generally positive tone, albeit not overwhelmingly so.

Subjectivity
Subjectivity: 0.4518 (approximately)
The subjectivity score measures how much personal opinion, emotion, or judgment is present in the text, versus factual information. Scores range from 0 (completely objective, or factual) to 1 (completely subjective). A score of 0.4518 indicates a moderate level of subjectivity. This suggests that the review contains a mix of personal opinion and factual content, leaning slightly more towards the objective/factual side.

Interpretation
Overall, the longest review in the dataset seems to express a positive sentiment with a balanced mix of personal opinions and objective facts. The positive polarity indicates that the reviewer likely had favorable things to say about the movie, while the moderate subjectivity score implies that the review is not solely based on personal feelings but includes some factual descriptions or assessments as well. This kind of sentiment and subjectivity combination can be typical for detailed movie reviews where the reviewer provides both their personal impression and some descriptive commentary on the movie's elements such as plot, acting, direction, etc.
